This dataset is a collection of Dungeons and Dragons 5e monster statistics sourced from Kaggle. The data was scraped from D&D database websites and contains all the basic monsters released in the original Monster Manual. Features in this dataset include the monster's name, challenge rating (CR), type, size, armor class (AC), hit points (HP), alignment, and core stats like Strength, Dexterity, and Constitution. This dataset is interesting to me because it allows us to use data science tools to analyze game balance and find the mathematical extremes of the monsters players face in a campaign.

```python
import pandas as pd 
df = pd.read_csv('dnd_monsters.csv')
```

```python
# Question 1: Which 5 monsters are the ultimate "tanks" with the highest Hit Points ?
highest_hp = df.sort_values(by='hp', ascending=False).head(5)
highest_hp.plot(x='name', y='hp', kind='bar', color='crimson', edgecolor='black', legend=False)
plt.title('Top 5 Monsters by Hit Points (HP)')
plt.xlabel('Monster Name')
plt.ylabel('Total HP')
plt.xticks(rotation=45)
plt.show()
# According to the data, the Tarrasque has the highest HP at 676, then Tiamat , the Zariel  followed by the Ancient Gold Dragon and the Ancient Red Dragon .
```

```python
# Question 2: What is the most common alignment for monsters in the D&D universe?
common_alignments = df['align'].value_counts().head(3)
common_alignments.plot(kind='pie', autopct='%1.1f%%', colors=['lightgray', 'purple', 'red', 'darkred', 'blue'])
plt.title('Top 5 Most Common Monster Alignments')
plt.ylabel('') 
plt.show()
# The most common alignment in this dataset is unaligned with 171 monsters. For monsters with a moral alignment
```

```python
# Question 3: How does a monster's Size affect its average Armor Class (AC)?
size_ac = df.groupby('size')['ac'].mean().sort_values(ascending=False)
size_ac.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Average Armor Class by Monster Size')
plt.xlabel('Monster Size')
plt.ylabel('Average AC')
plt.xticks(rotation=45)
plt.show()
# The Gargantuan monsters have the highest average Armor Class at 19.8, while Tiny monsters have the lowest average at 12.8.
```

```python
# Question 4: What are the top 3 most common types of monsters players will fight?
common_types = df['type'].value_counts().head(3)
common_types.sort_values().plot(kind='barh', color='forestgreen', edgecolor='black')
plt.title('Top 5 Most Common Monster Types')
plt.xlabel('Number of Monsters in Manual')
plt.ylabel('Monster Type')
plt.show()
# The most common monster types players will encounter in this manual are Beasts 106 , followed by monstrosity 75 and humanoid 68.
```

```python
# Question 5: Which monster has the absolute highest physical Strength score in the game?
max_str = df.sort_values(by='str', ascending=False).head(1)
plt.plot(max_str['name'], max_str['str'], marker='o', linestyle='', color='darkorange', markersize=10)
plt.title('Monsters with the Highest Strength Scores')
plt.xlabel('Monster Name')
plt.ylabel('Strength Score')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
# the Ancient Red Dragon has the highest strength
```
